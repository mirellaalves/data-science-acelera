### Feature engineering

Ao observar o Dicionário de Microdados do Enem-2016, verifica-se a seguinte informação:
"3. Foi considerado treineiro o inscrito que tinha menos de 18 anos no primeiro dia de realização do exame (05/11/2016) e que não havia concluído o ensino médio e não o concluiria em 2016"

Sendo assim, bastaria aplicar essa regra para o dataframe de teste.

```python
df_test['NU_IDADE'].head(20)
```

```python
df_test['TP_ST_CONCLUSAO'].head(20)
```

```python
# Reduzindo as features para apenas aquelas necessárias para o exercício
df_test_reduced = df_test[['NU_INSCRICAO', 'NU_IDADE', 'TP_ST_CONCLUSAO']]
df_test_reduced
```

```python
# Cria dataframe treineiro e não treineiros
df_test_treineiros = df_test_reduced.loc[(df_test_reduced['NU_IDADE'] < 18) & (df_test_reduced['TP_ST_CONCLUSAO'] > 2)]
df_test_no_treineiros = df_test_reduced.drop(df_test_treineiros.index)
df_test_treineiros
```

```python
# Exclui colunas 'NU_IDADE' e 'TP_ST_CONCLUSAO' dos dataframes
df_test_treineiros = df_test_treineiros.drop(['NU_IDADE', 'TP_ST_CONCLUSAO'], axis=1)
df_test_no_treineiros = df_test_no_treineiros.drop(['NU_IDADE', 'TP_ST_CONCLUSAO'], axis=1)
df_test_treineiros
```

```python
# Define o dataframe final

df_test_no_treineiros['IN_TREINEIRO'] = 0
df_test_treineiros['IN_TREINEIRO'] = 1

df_test_final = pd.concat([df_test_no_treineiros, df_test_treineiros])
df_test_final
```

```python
# Convertendo o dataframe em arquivo .csv
df_test_final.to_csv("answer-a.csv", index=False)
```
